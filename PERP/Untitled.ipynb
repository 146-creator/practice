```python
import pandas as pd
import requests
import numpy as np
import json
import pprint
import time

url = "https://api.97caijing.com/d1/issue/index.html"
url2 = 'https://api.97caijing.com/d1/bond/bond.html?'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.5112.81Safari/537.36 Edg/104.0.1293.54"}
category=[]
company = []
num = []
date = []
name = []
rate = []
aim = []
price = []
re_turn = []
date_right = []
date_right1 = []
for i in range(1013):
    i+=1
    data = {"page": i,
            "platform": "web"}
    try:
        response = requests.post(url = url,headers = headers,data = data)
        list_data = response.json()
        time.sleep(0.5)
        for i in range(len(list_data['data']["list"])):
            category.append(list_data["data"]['list'][i]['head']['category'])
            company.append(list_data["data"]['list'][i]['head']['list'][0]['value'])
            num.append(list_data["data"]['list'][i]['head']['list'][4]['value'])
            date.append(list_data["data"]['list'][i]['head']['list'][5]['value'])
            name.append(list_data["data"]['list'][i]['head']['title'])
            rate.append(list_data["data"]['list'][i]['list'][31]['value'])
            aim.append(list_data["data"]['list'][i]['list'][47]['value'])
            date_right.append(list_data["data"]['list'][i]['list'][43]['value'])
            date_right1.append(list_data["data"]['list'][i]['list'][43]['value'].split('、')[0])

            bond_id = str(list_data["data"]['list'][i]['head']['bond_id'])
            param = {
            'id': bond_id,
            'platform': 'web',
            'tradition_chinese': 'zh-cn',
                }
            response2 = requests.get(url = url2,headers = headers,params = param)
            data_list = response2.json()
            re_turn.append(data_list['data']['list']['price']['mid_yields'])
            price.append(data_list['data']['list']['price']['mid_price'])

    except Exception as e:
        pass
    continue



dataframe = pd.DataFrame([company,name,date,rate,price,re_turn,date_right1,date_right,num,category,aim,]).T
dataframe.to_csv('abc.csv',encoding='utf_8_sig')
```
